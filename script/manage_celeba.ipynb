```python
img_dir=r"E:\dataset\celeba\masked_celeba"
out_put_dir=r"E:\dataset\celeba\classed_celeba"
txt_file=r"E:\dataset\celeba\identity_CelebA.txt"
import os
```

```python
import shutil
with open(txt_file,'r',newline='\n') as identities:
    for line in identities.read().splitlines():
        img,id=line.split(' ')
        resource=os.path.join(img_dir,img)
        if not os.path.isfile(resource):
            print("not found ",img,id )
            continue
        destination=os.path.join(out_put_dir,id)
        if not os.path.exists(destination): os.mkdir(destination)
        shutil.move(resource,os.path.join(destination,img))
```
